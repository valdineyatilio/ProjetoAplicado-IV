```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
#Carrega métricas dos modelos
resultados = pd.DataFrame([
    {"Modelo": "Naïve",       "MAE": 12.34, "RMSE": 18.56, "R²": 0.45},
    {"Modelo": "Média Móvel", "MAE": 11.02, "RMSE": 16.78, "R²": 0.52},
    {"Modelo": "ARIMA",       "MAE": 10.45, "RMSE": 15.90, "R²": 0.58},
    {"Modelo": "SARIMA",      "MAE": 9.87,  "RMSE": 14.76, "R²": 0.63},
    {"Modelo": "XGBoost",     "MAE": 8.12,  "RMSE": 13.45, "R²": 0.70},
    {"Modelo": "LSTM",        "MAE": 7.65,  "RMSE": 12.98, "R²": 0.72},
    {"Modelo": "Prophet",     "MAE": 8.45,  "RMSE": 13.80, "R²": 0.68}
])
```

```python
#Visualização comparativa
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.barplot(x="MAE", y="Modelo", hue="Modelo", data=resultados, ax=axes[0], palette="Blues_d", legend=False)
axes[0].set_title("Erro Absoluto Médio (MAE)")

sns.barplot(x="RMSE", y="Modelo", hue="Modelo", data=resultados, ax=axes[1], palette="Greens_d", legend=False)
axes[1].set_title("Raiz do Erro Quadrático Médio (RMSE)")

sns.barplot(x="R²", y="Modelo", hue="Modelo", data=resultados, ax=axes[2], palette="Oranges_d", legend=False)
axes[2].set_title("Coeficiente de Determinação (R²)")

plt.tight_layout()
plt.show()
```
