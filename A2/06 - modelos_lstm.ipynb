```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Configurações
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
#Carregamendo do dataset 
df = pd.read_csv("data/processed/dataset_clean.csv")
print("Shape:", df.shape)
```

```python
#Seleciona variáveis
target = "casos"
features = [
    "Rt", "pop", "tempmed", "umidmed", "t_mean", "precip_sum", "pop_density",
    "casos_lag1", "casos_lag2", "casos_lag3", "casos_lag4", "casos_roll4"
]

X = df[features]
y = df[target]
```

```python
#Normaliza os dados
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))
```

```python
#Cria sequências para LSTM
def create_sequences(X, y, seq_len=4):
    X_seq, y_seq = [], []
    for i in range(seq_len, len(X)):
        X_seq.append(X[i-seq_len:i])
        y_seq.append(y[i])
    return np.array(X_seq), np.array(y_seq)

X_seq, y_seq = create_sequences(X_scaled, y_scaled, seq_len=4)
print("Sequências:", X_seq.shape)
```

```python
#Divide em treino e teste
split = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y_seq[:split], y_seq[split:]
```

```python
#Define modelo LSTM
from tensorflow.keras.layers import Input

model = Sequential()
model.add(Input(shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(64, activation="tanh"))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")
model.summary()
```

```python
#Treina modelo
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, verbose=0)
```

```python
#Avaliação
y_pred_scaled = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_true = scaler_y.inverse_transform(y_test)

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2 = r2_score(y_true, y_pred)

print(f"LSTM → MAE: {mae:.4f} | RMSE: {rmse:.4f} | R²: {r2:.4f}")
```

```python
#Visualização
plt.plot(y_true, label="Real")
plt.plot(y_pred, label="Previsto")
plt.title("Previsão de Casos - LSTM")
plt.legend()
plt.tight_layout()
plt.show()
```
